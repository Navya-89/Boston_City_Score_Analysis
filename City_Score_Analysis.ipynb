```python
import pandas as pd
import numpy as np
df_city_score = pd.read_csv('bostoncityscore.csv')
```

```python
df_city_score
```

```python
#@title
df_city_score['score_calculated_ts'] = pd.to_datetime(df_city_score['score_calculated_ts'] )
```

```python
#@title
df_city_score['year'] = df_city_score['score_calculated_ts'].dt.year
```

```python
#@title
df_subset = df_city_score[['metric_name', 'year', 'day_score']]
df_subset
```

```python
df_subset['year'].nunique()
```

```python
#@title
df_subset["status"] = np.where(df_subset["day_score"] >1 , 1, 0)
df_subset
```

```python
#@title
df_table_pivot = df_subset.groupby(['metric_name','year'],as_index=False)['status'].sum()
df_table_pivot
```

```python
#@title
df_result = df_table_pivot.pivot(index='metric_name', columns='year', values='status')
```

```python
#@title
df_result
```

```python
#@title
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(df_result, cmap="YlGnBu", linewidth=0.5, xticklabels=True, yticklabels=True)
plt.gcf().set_size_inches(10, 10, )
```
